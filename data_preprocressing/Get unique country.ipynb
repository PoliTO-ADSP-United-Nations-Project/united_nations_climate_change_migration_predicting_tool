## Greece distinct

```python
list_of_excel = [
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/CLEANED Q4 2017 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/CLEANED Q4 2018 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/CLEANED Q4 2019 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/CLEANED Q4 2021 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/Q4 2017 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/Q4 2018 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/Q4 2019 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/Q4 2020 Arrivals_to_Europe - GRC.xlsx",
    "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/GRC/Q4 2021 Arrivals_to_Europe - GRC.xlsx"
]
```

```python
my_ISO_list = []
for excel_file in list_of_excel:
  df = pd.read_excel(excel_file)
  df.columns = df.columns.str.replace(' ', '_')
  for iso in df['ISO3_of_Origin'].unique():
    my_ISO_list.append(iso)

# Remove duplicates
my_ISO_list = list(set(my_ISO_list))
```

```python
len(my_ISO_list)
```

```python
my_country = [(iso, pycountry.countries.get(alpha_3=iso)) for iso in my_ISO_list]
```

```python
none_list = []
for i in range(len(my_country)):
  country = my_country[i]
  if country[1] is None:
    none_list.append(i)

for index in none_list:
  print("Unknown: " + my_ISO_list[index])
```

```python
for country in my_country:
  if country[1] is not None:
    print(country[0] + '\t' + country[1].name)
  else:
    print(country[0] + '\t' + ' ')
```
