```python
my_ISO_list = ['OOS',
'MAR',
'DZA',
'MLI',
'BFA',
'GIN',
'SYR',
'EGY',
'CIV',
'GMB',
'LBR',
'LBY',
'SDN',
'SEN',
'TUN',
'PSE',
'SLE',
'BGD',
'CAF',
'ZAF',
'CMR',
'ERI',
'GNB',
'NGA',
'TCD',
'YEM',
'IRQ',
'LBN',
'TGO',
'BEN',
'GAB',
'NER',
'PAK',
'ETH']
```

```python
my_country = [pycountry.countries.get(alpha_3=iso) for iso in my_ISO_list]
```

```python
none_list = []
for i in range(len(my_country)):
  country = my_country[i]
  if country is None:
    none_list.append(i)

for index in none_list:
  print("Unknown: " + my_ISO_list[index])
  my_country[index] = ''
```

```python
for country in my_country:
  if isinstance(country, str):
    print(country)
  else:
    print(country.name)
```
