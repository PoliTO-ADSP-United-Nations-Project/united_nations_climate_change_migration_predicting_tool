```python
import pandas as pd

fileName = "2017_Yearly_Overview_Med_Crisis_0.xlsx"
inputPath = "./drive/MyDrive/datasets/" + fileName
outputPath = "./drive/MyDrive/datasets/"
df = pd.read_excel(inputPath, sheet_name="Origin Nationalites ITA-GRC-ESP", header=5, names=None, index_col=None, usecols=[0,1,2,7])

df.head()
```

```python
df_ita = df.loc[df['ISO3 of Arrival'] == "ITA"]
df_esp = df.loc[df['ISO3 of Arrival'] == "ESP"]
df_grc = df.loc[df['ISO3 of Arrival'] == "GRC"]
```

```python
origin_list = set(df_ita['ISO3 of Origin'])
origin_list
```

```python
df_ita["inflows"] = 0

for origin in origin_list:
  
  cumulate = 0

  for index, row in df_ita.iterrows():

    if row[1] == origin:
      
      trueValue = row[2] - cumulate
      cumulate = row[2]
      df_ita.loc[index,"inflows"] = trueValue

df_ita.head(20)
```
