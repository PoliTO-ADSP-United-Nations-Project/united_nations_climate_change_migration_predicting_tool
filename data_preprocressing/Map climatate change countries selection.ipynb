```python
import matplotlib.pyplot as plt
import geopandas
import pandas as pd
```

```python
file_of_countries = "/content/drive/MyDrive/University /Polito Master/Applied data science project/Project/datasets/Download climate dataset - Distinct country.xlsx"
```

```python
df = pd.read_excel(file_of_countries)
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
ax = world.plot()

for country in df['Country'].values:
  if country != "":
    world[world.name == country].plot(color='yellow',ax=ax)

plt.show()
```
