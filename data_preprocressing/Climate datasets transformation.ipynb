## Cleaning/Transformation


```python
def clean(df):
    df.drop("Unnamed: 0",axis=1,inplace=True)
    dates=[]
    values=[]
    months=df.columns.tolist()
    months.remove("Year")
    for year in range(1901,2021):
        for month in range(1,13):
            dates.append(pd.to_datetime(str(year)+"-"+str(month)))
    for year in range(1901,2021):
        for month in months :
            values.append(df.loc[df["Year"]==year][month].values[0])
    df_new= pd.DataFrame(list(zip(dates,values)),columns =['ds', 'y'])
    return df_new
```

```python
initial_path= "/content/drive/MyDrive/Climate datasets"
internal_folders=["Max temperature","Mean temperature","Min temperature","Percipitation"]
paths=[]
for folder in internal_folders:
    paths.append(initial_path+"/"+folder)
paths
```

```python
cleaned_datasets=[]
file_names=[]
for path in paths:
    for file_name in os.listdir(path):
        with open(os.path.join(path, file_name)) as fr:
                print(os.path.join(path,file_name))
                df=pd.read_csv(fr)
                file_names.append(os.path.join(path,file_name))
                cleaned_datasets.append(clean(df))
```

## File saving


```python
for i,(dataset,name) in enumerate(zip(cleaned_datasets,file_names)):
    fp=name.replace("Climate datasets","Climate_datasets_Cleaned")
    dir=fp[:-15]
    
    os.chdir(dir)  
    dataset.to_csv(name.split("/")[-1])
```
